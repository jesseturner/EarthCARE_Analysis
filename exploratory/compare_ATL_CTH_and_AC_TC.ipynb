## Comparing between ATL_CTH_2A and AC__TC__2B

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta
import matplotlib.dates as mdates
from matplotlib.colors import ListedColormap
import matplotlib.colors as mcolors
```

```python
class_path = 'data_AC__TC__2B/ECA_EXAC_AC__TC__2B_20250525T234738Z_20250526T014041Z_05633D.h5'
cth_path = 'data_ATL_CTH_2A/ECA_EXAE_ATL_CTH_2A_20250525T234738Z_20250526T014141Z_05633D.h5'
```

```python
with open('structure_AC__TC__2B.txt', 'w') as out_file:
    with h5py.File(class_path, 'r') as f:

        def print_structure(name, obj):
            if isinstance(obj, h5py.Group):
                out_file.write(f"[Group] {name}\n")
            elif isinstance(obj, h5py.Dataset):
                out_file.write(f"[Dataset] {name} - shape: {obj.shape}, dtype: {obj.dtype}\n")

            for key, val in obj.attrs.items():
                out_file.write(f"    [attr] {key}: {val}\n")

        f.visititems(print_structure)

        if f.attrs:
            out_file.write("\n[File-level Attributes]\n")
            for key, val in f.attrs.items():
                out_file.write(f"    [attr] {key}: {val}\n")

```

```python
with open('structure_ATL_CTH_2A.txt', 'w') as out_file:
    with h5py.File(cth_path, 'r') as f:

        def print_structure(name, obj):
            if isinstance(obj, h5py.Group):
                out_file.write(f"[Group] {name}\n")
            elif isinstance(obj, h5py.Dataset):
                out_file.write(f"[Dataset] {name} - shape: {obj.shape}, dtype: {obj.dtype}\n")

            for key, val in obj.attrs.items():
                out_file.write(f"    [attr] {key}: {val}\n")

        f.visititems(print_structure)

        if f.attrs:
            out_file.write("\n[File-level Attributes]\n")
            for key, val in f.attrs.items():
                out_file.write(f"    [attr] {key}: {val}\n")
```

```python
#--- data from cloud top height
with h5py.File(cth_path, 'r') as f:
    dataset_path = 'ScienceData/ATLID_cloud_top_height'
    cth_data = np.array(f[dataset_path][()])

    time_path = 'ScienceData/time'
    time_data = np.array(f[time_path][()])

    consistency_path = 'ScienceData/ATLID_cloud_top_height_consistency'
    consistency_data = np.array(f[consistency_path][()])
```

```python
mask = cth_data != 9.96921e+36
cth = cth_data[mask]/1000 # convert to km

ref_time = datetime(2000, 1, 1, 0, 0, 0)
cth_time = [ref_time + timedelta(seconds=sec) for sec in time_data[mask]]
```

```python
#--- data from cloud classification
with h5py.File(class_path, 'r') as f:
    dataset1_path = 'ScienceData/synergetic_target_classification'
    class_data = np.array(f[dataset1_path][()])

    dataset2_path = 'ScienceData/height'
    class_height_data = np.array(f[dataset2_path][()])

    time_path = 'ScienceData/time'
    class_time_data = np.array(f[time_path][()])
```

```python
class_height = class_height_data/1000 # convert to km

ref_time = datetime(2000, 1, 1, 0, 0, 0)
class_time = [ref_time + timedelta(seconds=sec) for sec in class_time_data]
```

```python
time_grid, height_grid = np.meshgrid(class_time, np.arange(class_height.shape[1]), indexing='ij')
```

```python
custom_colors = [
    '#c5c9c7', '#a2653e', '#ffffff', '#ff474c', '#0504aa', '#009337', '#840000',
    '#042e60', '#d8dcd6', '#ffff84', '#f5bf03', '#f97306', '#ff000d', '#5539cc',
    '#2976bb', '#0d75f8', '#014182', '#017b92', '#06b48b', '#aaff32', '#6dedfd',
    '#01f9c6', '#7bc8f6', '#d7fffe', '#a2cffe', '#04d9ff', '#7a9703', '#b2996e',
    '#ffbacd', '#d99b82', '#947e94', '#856798', '#ac86a8', '#59656d', '#76424e',
    '#363737'
]
```

```python
plt.figure()
fig, ax = plt.subplots(figsize=(9, 4))

cmap = ListedColormap(custom_colors)
pcm = ax.pcolormesh(time_grid, class_height, class_data, cmap=cmap, shading='auto')

ax.scatter(cth_time, cth, color='red', marker='x', s=2, label="Cloud Top Height")

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
fig.autofmt_xdate()
ax.legend()

ax.set_title("")
ax.set_xlabel("Time")
ax.set_ylabel("Height (km)")

plt.savefig('compare_cth_to_class.png', dpi=300)
plt.show()
```

```python
sentinel = -99

no_clouds_both = consistency_data.copy()
mask = no_clouds_both[:, 0] != 0
no_clouds_both[mask] = sentinel # replace values so it stays consistent with time

no_clouds_a_tc = consistency_data.copy()
mask = no_clouds_a_tc[:, 0] != 1
no_clouds_a_tc[mask] = sentinel

no_clouds_a_cth = consistency_data.copy()
mask = no_clouds_a_cth[:, 0] != 2
no_clouds_a_cth[mask] = sentinel

clouds_both = consistency_data.copy()
mask = clouds_both[:, 0] != 3
clouds_both[mask] = sentinel
```

```python
counts = [
    np.sum(no_clouds_both != -99),
    np.sum(no_clouds_a_tc != -99),
    np.sum(no_clouds_a_cth != -99),
    np.sum(clouds_both != -99)
]

labels = ['No clouds in both', 'No clouds in A-TC', 'No clouds in A-CTH', 'Clouds in both']

plt.figure(figsize=(6, 4))
plt.bar(labels, counts, color='skyblue')
plt.xticks(rotation=45)
plt.ylabel('Count of Valid Values')
plt.title('Comparison between CBH and TC')
plt.tight_layout()
plt.show()
```

```python
class_time = np.array(class_time)
consistency_values = consistency_data[:, 1]
```

```python
custom_colors = [
    '#c5c9c7', '#a2653e', 'None', '#ff474c', '#0504aa', '#009337', '#840000',
    '#042e60', '#d8dcd6', '#ffff84', '#f5bf03', '#f97306', '#ff000d', '#5539cc',
    '#2976bb', '#0d75f8', '#014182', '#017b92', '#06b48b', '#aaff32', '#6dedfd',
    '#01f9c6', '#7bc8f6', '#d7fffe', '#a2cffe', '#04d9ff', '#7a9703', '#b2996e',
    '#ffbacd', '#d99b82', '#947e94', '#856798', '#ac86a8', '#59656d', '#76424e',
    '#363737'
]
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))

# --- Left y-axis: pcolormesh + scatter ---
cmap1 = ListedColormap(custom_colors)
pcm = ax1.pcolormesh(time_grid, class_height, class_data, cmap=cmap1, shading='auto')
ax1.scatter(cth_time, cth, color='red', marker='x', s=10, label="Cloud Top Height")

ax1.set_ylabel("Height (units)")
ax1.set_xlabel("Time")
ax1.set_title("Combined Cloud Height and Consistency")

# Format x-axis dates
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
fig.autofmt_xdate()

# Add colorbar for pcolormesh
cbar1 = fig.colorbar(pcm, ax=ax1, orientation='vertical', pad=0.02)
cbar1.set_label('Class Data')

# --- Right y-axis: vertical lines for consistency values ---
ax2 = ax1.twinx()  # share x-axis
ax2.set_zorder(0)  # Send ax2 to the back
ax1.set_zorder(1)  # Ensure ax1 is in front
ax2.patch.set_visible(False) # Make ax2 background transparent
ax1.patch.set_visible(False)

# Prepare colormap for consistency values 0-10
cmap2 = plt.get_cmap('Greens', 11)
norm2 = mcolors.BoundaryNorm(boundaries=np.arange(0, 10, 1), ncolors=11)

# Filter out invalid values if needed (optional)
valid_mask = (consistency_values >= 0) & (consistency_values <= 10)
times_valid = class_time[valid_mask]
values_valid = consistency_values[valid_mask].astype(int)

# Plot vertical lines with color by consistency value
for t, val in zip(times_valid, values_valid):
    ax2.vlines(t, ymin=0, ymax=1, color=cmap2(norm2(val)), linewidth=1)

ax2.set_ylim(0, 1)
ax2.set_ylabel('')

# Optional: add a colorbar for consistency values
from matplotlib.cm import ScalarMappable
sm = ScalarMappable(cmap=cmap2, norm=norm2)
sm.set_array([])
cbar2 = fig.colorbar(sm, ax=ax2, orientation='vertical', pad=0.12)
cbar2.set_label('Consistency Value (0-10)')

# Optional legend for scatter points (on ax1)
ax1.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
